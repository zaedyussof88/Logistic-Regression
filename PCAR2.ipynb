```python
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # Required for 3D plotting
```

```python
# Step 1: Create a sample multivariate dataset
np.random.seed(42)
data = np.random.rand(50, 3)  # 50 samples with 3 features each
```

```python
data
```

```python
# Step 2: Standardize the data (optional, but recommended)
mean = np.mean(data, axis=0)
std_dev = np.std(data, axis=0)
standardized_data = (data - mean) / std_dev
```

```python
standardized_data
```

```python
# Step 3: Fit the PCA model
pca = PCA()
pca.fit(standardized_data)

# Access eigenvalues and eigenvectors
eigenvalues = pca.explained_variance_
eigenvectors = pca.components_

# Access covariance matrix
covariance_matrix = pca.get_covariance()

# Calculate cumulative percentage of explained variance
cumulative_explained_variance = np.cumsum(pca.explained_variance_ratio_)

# Find the number of principal components that explain 95% of the variance
num_components_100_percent = np.argmax(cumulative_explained_variance >= 1.0) + 1

# Display the results
print("Eigenvalues:")
print(eigenvalues)

print("\nEigenvectors:")
print(eigenvectors)

print("\nCovariance Matrix:")
print(covariance_matrix)

print("\nExplained Variance Ratio:")
print(pca.explained_variance_ratio_)

print("\nCumulative Explained Variance:")
print(cumulative_explained_variance)

print(f"\nNumber of Principal Components for 100% Variance: {num_components_100_percent}")
```

```python
# Step 4: Transform the data to the new coordinate system
transformed_data = pca.transform(standardized_data)
```

```python
transformed_data
```

```python
# Step 5: Plot the data in the new coordinate system
if data.shape[1] == 2:
    plt.scatter(transformed_data[:, 0], transformed_data[:, 1])
    plt.title('PCA: Transformed Data')
    plt.xlabel('Principal Component 1')
    plt.ylabel('Principal Component 2')
    plt.show()
elif data.shape[1] == 3:
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.scatter(transformed_data[:, 0], transformed_data[:, 1], transformed_data[:, 2])
    ax.set_title('PCA: Transformed Data')
    ax.set_xlabel('Principal Component 1')
    ax.set_ylabel('Principal Component 2')
    ax.set_zlabel('Principal Component 3')
    plt.show()
else:
    print("Data has more than 3 features. Visualization not supported.")
```

```python
# Step 6: Plot the explained variance ratio
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.title('Explained Variance Ratio')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.show()
```
